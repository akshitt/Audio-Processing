## Coding a Sinusoid

```python
#libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple Sinewave
A = 1         #amplitude
f0 = 1000     #frequency
phi = np.pi/2 #phase
fs = 44100    #sampling rate
t = np.arange(-0.002, 0.002, 1.0/fs)
x = A*np.cos(2*np.pi*f0*t + phi)
plt.plot(t,x)
plt.axis([-.002,.002,-1,1])
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
# Complex Sinewave
N = 1000 #number of samples
k = 4 #number of periods
n = np.arange(-N/2,N/2)

s = np.exp(2*np.pi*1j*k*n/N)
plt.plot(n, np.imag(s))
plt.plot(n, np.real(s))
plt.legend(['Imaginary','Real'])
```

## DFT on Different Sinusoids

```python
#Discrete Fourier Transform of Complex Signal

#input wave
N = 70
k0 = 23
x = np.exp(2*np.pi*1j*k0/N * np.arange(N))   

#output wave
X = np.array([])
for k in range(N):
    s = np.exp(-2*np.pi*1j*k/N * np.arange(N))
    X = np.append(X, sum(x*s))
    
plt.plot(np.arange(N),abs(X))
```

```python
#Discrete Fourier Transform of Real Signal

#input wave
N = 69
k0 = 28
x = np.cos(2*np.pi*k0/N * np.arange(N))   

#output wave
X = np.array([])
for k in range(N):
    s = np.exp(-2*np.pi*1j*k/N * np.arange(N))
    X = np.append(X, sum(x*s))
    
plt.plot(np.arange(N),abs(X))

#peaks at 28 & 41 (69-28)
```

```python
#When k is not int
#input wave
N = 69
k0 = 24.7
x = np.cos(2*np.pi*k0/N * np.arange(N))   

#output wave
X = np.array([])
for k in range(N):
    s = np.exp(-2*np.pi*1j*k/N * np.arange(N))
    X = np.append(X, sum(x*s))
    
plt.plot(np.arange(N),abs(X))
```

## Inverse DFT

```python
#Inverse DFT

#input wave
N = 256
k0 = 27
x = np.cos(2*np.pi*k0/N * np.arange(N))   

#output wave
X = np.array([])
for k in range(N):
    s = np.exp(-2*np.pi*1j*k/N * np.arange(N))
    X = np.append(X, sum(x*s))

#getting input wave back
y = np.array([])
for q in range(N):
    s = np.exp(2*np.pi*1j*q/N * np.arange(N))
    y = np.append(y, 1.0/N*sum(X*s))

plt.plot(np.arange(N),x)
plt.plot(np.arange(N),y)
plt.legend(['Input','Generated'])
#Complete overlap obviously
```
